```python
x = "hello"
y = x

y = y + " world"

print(x)
print(y)

# Why does x remain unchanged?  beacuse the strings are immutable so if we copy the strings x=y it will point to same refernce but when we modify the y it will point to new reference hence x is unchanged
# 
a = 10
b = a

b = 20

print(a) # answer is 10 since int is immutable

a = "cat"
b = a

b += "dog"

print(a) # answer is cat
```

```python
a = [1,2]
b = [3,4]

a.append(b)

print(a) # [1,2,[3,4]]

a = [1,2]
b = [3,4]

a.extend(b)

print(a) # difference is that append adds the list into the list but extends extends the list a with the elements of list b

# reverse without using reverse()
a = [1,2,3,4]
print(a[::-1])

# find the max element without using max()
nums = [3,7,2,9,1]
i=0
while i<len(nums)-1:
    if(nums[i]>nums[i+1]):
        max = nums[i]
    i += 1
print(max)

#remove duplicates 
a = [1,1,2,2,3,4,4]
b = []

for i in a:
    if i not in b:
        b.append(i)
print(b) 
```
