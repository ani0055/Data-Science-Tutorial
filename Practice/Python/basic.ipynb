```python
# 1. Reverse list without reverse()

l = ['A', "25", 16, True, "abc"]
i = 0
j = len(l)-1
while(i<j):
    l[i],l[j] = l[j],l[i]
    i += 1
    j -= 1

print(l)


# 2. Find second largest element
l = [1,14,25,62,5,48,61,62]

i = 0
max = 0
max2 = 0

while(i<len(l)-1):
    if(l[i]>max2):
        if(l[i]>=max):
            max2 = max
            max = l[i]
        else:
            max2 = l[i]
    i += 1
        
print(max2)


# 3. Check palindrome string
str = "hannah"
if(str == str[::-1]):
    print(True)
else:
    print(False)

# 4. Count vowels
str = "My name is Khan"
vow = "aeiouAEIOU"
count = 0

for char in str:
    if char in vow:
        count += 1
print(count)
# 5. Frequency of characters
str = "apple banana strawberry"
freq = {}

for char in str:
    freq[char] = freq.get(char,0)+1

print(freq)

# 6. Remove duplicates from list
l = ['a','b','c','a','d','b','d','e']
l1 = set({})

for i in l:
    l1.add(i)

print(l1)

# 7. Find duplicate elements
l1 = set({})
l2  = set({})
for i in l:
    if i in l1:
        l2.add(i)
    else:
        l1.add(i)
print(l2)
# 8. Merge two dictionaries
d1 = {1: "a", 2:"b", 3:"c"}
d2 = {4: "d", 5:"e", 6:"f"}

merged = d1|d2
print(merged)

# 9. Invert dictionary
original = {1: 'a', 2: 'b'}
inverted = {value: key for key, value in original.items()}
print(inverted)

# 10. Sort list manually (without sort())
l = [1,4,6,4,5,7,9]
for  i in range (0,len(l)):
    for j in range(i,len(l)-1):
        if(l[j]>l[j+1]):
            l[j],l[j+1] = l[j+1],l[j]
print(l)
```
