```python
import numpy as np
```

```python
arr = np.array([[10,20,30,40],[50,60,70,80]])
print(arr)
print(type(arr))
```

```python
# Slicing of an Array
a = np.array([10,12,15,18,28])
print(a[2:])
print(a[1:4])
print(a[:3])
```

```python
# Slicing of an multi dimensional array

array = np.array([[16,26,36,46],[58,61,98,72]])
print(array[:2, 1:4])
print(array[1,3]) # for accessing the elements of idividual array
```

```python
print(np.shape(arr)) # gives the shape of the array no. of rows and coloum
print(np.size(arr))  # gives the total number of elements in the array
print(np.ndim(arr))  # Gives the dimensions of the array
print(arr.dtype)     # Gives the datatype of the array
```

```python
# Inspecting an array
# a.shape - array dimension
# len(a) - length of array
# a.ndim - number of array dimensions
# a.size - number of array elements
# a.dtype - Data type of array elements
# a.astype(int) - convert on array to different type

import numpy as np

a = [30,40,60,50]
arr = np.array(a)
print(arr)
print(len(arr))
print(np.size(arr))
print(type(arr))
print(arr.dtype)
print(arr.astype(float))
```

```python
#Fuctions on an array

arr1 = np.array([20,30,42,58,69])
arr2 = np.array([26,48,59,76,81])
print(arr1+arr2)
print(np.add(arr1,arr2))

print("-----------------------------------------------")
print(arr1-arr2)
print(np.subtract(arr1,arr2))

print("-----------------------------------------------")
print(arr1*arr2)
print(np.multiply(arr1,arr2))

print("-----------------------------------------------")
print(arr1/arr2)
print(np.divide(arr1,arr2))
```

```python
arr1 = np.array([3,4,6,8,11,32,48])
arr2 = np.array([4])
print(np.power(arr1,arr2))
print(np.pow(arr1,2))

arr3 = np.array([25,36,81,225,576])
print(np.sqrt(arr3))
```

```python
# Combining ans Spliting in array

# Concatenate

arr1 = np.array([36,41,58])
arr2 = np.array([96,54,88,48])
print(np.concatenate([arr1,arr2]))

```

```python
arr1 = np.array([[36,41,58], [91,87,56]])

arr2 = np.array([[96,54,88],[48,66,59]])



# Concatenation  on the basis of axis
print(np.concatenate([arr1,arr2],axis=0))
print()
print(np.concatenate([arr1,arr2],axis=1))
print()

# Horizontal concatenation
print(np.hstack([arr1,arr2]))
print()

#Vertical Ci=oncatenation
print(np.vstack([arr1,arr2]))
print()
```

```python
a = np.array([21,63,59,74,12,45,86,99,23])
b = np.array_split(a,3) #Splits array into 3 parts
c = np.array_split(a,2)
print(b)
print(b[2])
print(c)
```

```python
# Adding and Removing Elements from the Data
# np.append(a,d) appends items in an array
# np.insert(a,1,5) inserts items in an array
# np.delete(a,[1]) delete items from an array

a = np.array([11,12,51,56,85])
b = np.array([[23,34,38],[51,68,74]])
print(np.append(a,[92,101]))

print(np.insert(a,2,50)) # array, index, value
print(np.insert(b,1,[50,60,70], axis = 0))

print(np.delete(a,2))
```

```python
# Sort, Filter, and Search

arr = np.array([56,42,15,41,23,2,10,11])
print(np.sort(arr))

s = np.where(arr%2 == 0) # find elemnts which are even
print(s)

fa = [True, False, True, False, True, False, True, False]
new = arr[fa]
print(new) # Elements that are True will be printed

fa = arr>40
print(arr[fa])
```

```python
# Aggregating an Array

a = np.array([45,25,65,85,95,75,15,25,35,55])

print(np.sum(a))
print(np.min(a))
print(np.max(a))
print(np.mean(a)) # gives the mean of the array
print(np.cumsum(a)) # gives the cumulative sum of the ARRAY
```

```python
Prices = np.array([100, 199, 120, 330, 520, 12])
Quantity = np.array([6, 3, 8, 2, 1, 20])

print(np.cumprod([Prices, Quantity], axis = 0)) #Gives price*quantity
fin = np.cumprod([Prices, Quantity], axis = 0)
print(fin[1].sum())
```

```python
import numpy as np
import statistics as stats

baked_food = [200,160,150,192,150,148,189,254,140]
a = np.array(baked_food)
print(np.mean(a)) # Average
print(np.median(a))
print(stats.mode(a))
print(np.std(a))
print(np.var(a)) #Square of Standard Deviation
```

```python
# -1 represents inveresly propotional relation
# 1 represents proportional relation
# 0 represents no relation
Tobacco_Consumption = [30,25,45,62,35,85,15]
deaths = [120,100,200,342,132,440,76]

print(np.corrcoef([Tobacco_Consumption, deaths]))
```
