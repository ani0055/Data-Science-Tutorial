```python
# Merge, Join and concatenate in Pandas
import pandas as pd

Data1 = {"Emp_Id": ["E01","E02","E03","E04","E05","E06"],
        "Names":["Ram","Shyam","Rahul","Vishal","Ravi","John"],
        "Age":[34,56,23,44,32,36]}

Data2 = {"Emp_Id":["E01","E02","E03","E04","E05","E06"],
        "Salary":[45000,60000,75000,30000,45000,80000]}

df1 = pd.DataFrame(Data1)
df2 = pd.DataFrame(Data2)

print(df1)
print()
print(df2)
```

```python
# Merge

print(pd.merge(df1,df2,on = "Emp_Id")) # merged on the basis of Emp_Id
print()

#Join
print(pd.merge(left = df1,right = df2,on = "Emp_Id",how = "left")) # Left Join
print()
print(pd.merge(left = df1,right = df2,on = "Emp_Id",how = "right")) #Right Join
print()
print(df1.join(df2.set_index("Emp_Id"), on="Emp_Id")) #Full Join
```

```python
# Concatenation 

Data1 = {"Emp_Id": ["E01","E02","E03","E04","E05","E06"],
        "Names":["Ram","Shyam","Rahul","Vishal","Ravi","John"],
        "Age":[34,56,23,44,32,36]}

Data3 = {"Emp_Id": ["E07","E08","E09","E10","E11","E12"],
        "Names":["Patrik","Rayan","Dave","Vicky","Romeo","Jake"],
        "Age":[26,41,32,49,35,52]}

df1 = pd.DataFrame(Data1)
df3 = pd.DataFrame(Data3)

print(pd.concat([df1,df3]))
```

```python
# Compare the DataFrames

import pandas as pd

dict = {"Fruits": ["Apples","Mangos","Banana","Papaya"],
       "Price":[150,100,50,40],
       "Quantity":[10,18,9,4]}

df = pd.DataFrame(dict)
print(df)
print()
df2 = df.copy()
df2.loc[0,"Price"]=120
df2.loc[1,"Price"]=175
df2.loc[3,"Price"]=30
df2.loc[0,"Quantity"]=12
df2.loc[1,"Quantity"]=6
df2.loc[3,"Quantity"]=5

print(df2)
print()
print(df.compare(df2))
print()
print(df.compare(df2,keep_shape=True)) # Gives alll the values of the data columns
```

```python
# Pivoting and Melting in DataFrame

import pandas as pd

Data1 = {"Keys":["k1", "k2", "k1", "k2"],
        "Names":["John", "Ben", "David", "Peter"],
        "Houses":["Red", "Blue", "Green", "Red"]}

df = pd.DataFrame(Data1)
print(df)

print(df.pivot(index = "Keys", columns="Names", values="Houses"))
print()

Data2 = {"Names":["John", "Ben", "David", "Peter"],
        "Houses":["Red", "Blue", "Green", "Red"],
        "Grades":["3rd", "8th", "9th", "8th"]}

df2 = pd.DataFrame(Data2)
print(df2)
print()
print(pd.melt(df2, id_vars=["Names"], value_vars=["Houses", "Grades"]))
```
