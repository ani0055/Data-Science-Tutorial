```python
# bar plot
import matplotlib.pyplot as plt

x = [98,67,88,95,88]
y = ["Part 1", "Part 2", "Part 3", "Part 4", "Part 5"]
plt.bar(y , x, color = "green") #For plotting a bar graph (use shift+tab+"+" to see the various attributes of the plot)
plt.xlabel("Parts of Book") #Labelling x and y axis
plt.ylabel("Popularity")
plt.title("Popularity of the Books") #giving heading to the graph
plt.show()
```

```python
# Line Plot mainly used for seeing frequency and fluctuations
import matplotlib.pyplot as plt

x =["day 1", "day 2", "day 3", "day 4", "day 5"]
y1 = [300, 420,250,280,400]
y2 = [320,380,360,500,420]


plt.plot(x,y1, marker = "o", color="red",label = "week1")
plt.plot(x,y2, marker = "*", color="green",label = "week2")# for ploting line plot
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend() # makes the box denoting the box in the axis which shows the values
plt.show()
```

```python
# Pie Chart
import matplotlib.pyplot as plt

brands=["OnePlus", "Apple", "Samsung", "Vivo", "Redmi", "Oppo"]
x = [20,36,42,18,27,16]
c = ["red", "blue", "magenta", "green", "orange", "yellow"]
ex = [0,0,0.05,0,0,0] # Explode separates the value from pie chart by 0.1 units used to highlight

plt.pie(x, labels=brands, colors = c, explode = ex, autopct = "%.2f", startangle = 180) #Plotting the pie chart for value of x to the brands 
                                                                                        # autopct gives us the percent values of the each sector of pie
                                                                                        # Startangle rotates the pie chart by the degree's mentioned
                                                                                        # also we can add shadow to the pie chart by "shadow = true"
plt.show()
```

```python
# Box Plot
# min(lower fence) and max(upper fence) values, Q1, median, Q3, where Q1 and Q3 are the quartiles 1 and 3 i.e., 25 and 75 %ile
import matplotlib.pyplot as plt

l = [25,46,57,25,45,13,29,39,46,99,22,12,14,16,55,6,46,52,55,78]
plt.boxplot(l)
plt.show()
```

```python
# SubPlots

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [45,34,56,23,45]

plt.subplot(1,2,1)
plt.title("Age")
plt.plot(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]
plt.subplot(1,2,2) # plots the 2 graph together on same axis where 1 is axis, 2 is no. of plots and 2 is the position of the plot
plt.title("Weight")
plt.plot(x,y)
plt.suptitle("Some Data") #Super Title for the complete graph
plt.show()
```
